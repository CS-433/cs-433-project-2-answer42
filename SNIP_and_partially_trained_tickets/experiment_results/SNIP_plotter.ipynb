```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = {
    'reference points': [0.4, 0.6, 0.8, 0.9, 0.95, 0.98],
    'ResNet32': {
        'original': [94.42, 94.72, 93.84, 92.55, 90.98, 87.49],
        'layerwise rearrange': [94.49, 94.29, 93.61, 92.72, 91.35, 87.62],
        'corrupted data': [94.39, 94.11, 93.90, 92.61, 90.51, 87.63]
    },
    'VGG19': {
        'original': [93.40, 93.48, 93.22, 93.29, 93.13, 89.56],
        'layerwise rearrange': [93.23, 93.28, 93.02, 93.32, 93.20, 88.41],
        'corrupted data': [93.02, 93.19, 93.57, 93.37, 92.71, 49.13]
    }
}
```

```python
def plot(ax, x, data):
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.plot(x, data['original'][:len(x)])
    ax.plot(x, data['layerwise rearrange'][:len(x)])
    ax.plot(x, data['corrupted data'][:len(x)])
    ax.tick_params(axis=u'both', which=u'both', direction='in')
    plt.axis([0.4, 1, 87.1, 94.99])
    plt.xlabel('Sparsity ratio')
    plt.ylabel('Test Accuracy')
    plt.grid()
    plt.show()
    
plot(
    plt.subplot(111),
    data['reference points'], 
    data['ResNet32'])

plot(
    plt.subplot(111),
    data['reference points'][:-1], 
    data['VGG19'])
```
